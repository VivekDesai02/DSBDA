```python
df=pd.DataFrame(dict1)
df
```

```python
col1=['Maths Score','Reading_Score','Writing_Score']
df.boxplot(col1)
```

```python
rscore=df['Reading_Score']
q1=np.percentile(rscore,25)
q3=np.percentile(rscore,75)
print(q1,q3)
```

```python
iqr=q3-q1
print(iqr)
```

```python
l_bound=q1-1.5*iqr
u_bound=q3+1.5*iqr
print(l_bound,u_bound)
```

```python
r_outlier=[]
for i in rscore:
    if(i<l_bound or i>u_bound):
        r_outlier.append(i)
print(r_outlier)
```

```python
median=np.median(rscore)
median
```

```python
df['Reading_Score']=np.where(df['Reading_Score']>u_bound,median,df['Reading_Score'])
```

```python
df['Reading_Score']
```

```python
median=np.median(df['Maths Score'])
median
```

```python
df['Maths Score']=np.where(df['Maths Score']>u_bound,median,df['Maths Score'])
df['Maths Score']
```

```python
df['Maths Score']=np.where(df['Maths Score']<l_bound,median,df['Maths Score'])
df['Maths Score']
```
